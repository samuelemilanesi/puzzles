```python
grid=[[6,2,1,3,6,1,7,7,4,3],
[2,3,4,5,7,8,1,5,2,3],
[1,6,1,2,5,1,6,3,6,2],
[5,3,5,5,1,6,7,3,7,3],
[1,2,6,4,1,3,3,5,5,5],
[2,4,6,6,6,2,1,3,8,8],
[2,4,0,2,3,6,5,2,4,6],
[3,1,7,6,2,3,1,5,7,7],
[6,1,3,6,4,5,4,2,2,7],
[6,7,5,7,6,2,4,1,9,1]]

min_len = 10000000
```

```python
def search_next(xs,ys):
    global grid
    res=[]
    for x in range(0,10):
        if x!=xs and abs(xs-x)==grid[ys][x]:
            res.append([x,ys])
    for y in range(0,10):
        if y!=ys and abs(ys-y)==grid[y][xs]:
            res.append([xs,y])
        
    if res == []: 
        grid[ys][xs]==0
        
    return res


def solve(x_start, y_start, cur_len, depth=0):
    print(str(x_start)+str(y_start))
    global min_len
    next = search_next(x_start,y_start)
    cur_len += grid[y_start][x_start]
    if(y_start==9 and x_start==0):
        if cur_len<min_len:
            min_len = cur_len
            print('ok'+str(min_len))
        return cur_len
    if depth>15:
        return
    for el in next:
        solve(el[0],el[1],cur_len,depth+1)



    
```

```python
solve(6,2,0)
```
